```python
location = pd.read_csv("Attandance.csv", encoding='latin1' )
```

```python
location.head(254)
```

```python
location['new_indicator']=location['Text'].str.contains('live location shared')
location.head(75)
```

```python
location = location[location['new_indicator'] == True]
```

```python
location
```

```python
Date_location= location["Text"].str.split(",", n = 1, expand = True)
```

```python
location1=Date_location.rename(columns={0:'Date',1:'Text'})
```

```python
location1
```

```python
location_date= location1["Date"].str.split("/", n = 2, expand = True)
```

```python
location_date=location_date.rename(columns={0:'Day',1:'Month',2:'Year'})
```

```python
location_date
```

```python
location_time= location1["Text"].str.split("-", n = 1, expand = True)
```

```python
location_time=location_time.rename(columns={0:'Time',1:'Text'})
```

```python
location_time
```

```python
location_time_pers= location_time["Text"].str.split(":", n = 1, expand = True)
```

```python
location_time_pers=location_time_pers.rename(columns={0:'person',1:'text'})
```

```python
location_time_pers
```

```python
location_time_space= location_time["Time"].str.split(" ", n = 1, expand = True)
```

```python
location_time_space=location_time_space.rename(columns={0:'Space',1:'Time'})
```

```python
location_time_space
```

```python
location_time_zone= location_time_space["Time"].str.split(" ", n = 1, expand = True)
```

```python
location_time_zone=location_time_zone.rename(columns={0:'Time',1:'zone'})
```

```python
location_time_zone
```

```python
location_time_pers
```

```python
result = pd.concat([location_date, location_time_zone, location_time_pers ], axis=1, join="inner")
```

```python
result
```

```python
#result['Datetime']=result[['Month','Day','Year']].astype(str).apply('-'.join,1)+' '+result['Time']
```

```python
result['Time'] = pd.to_datetime(result['Time'],format= '%H:%M' ).dt.time
```

```python
result['Datetime']=result[['Month','Day','Year']].astype(str).apply('/'.join,1)
```

```python
result=result.drop(columns=['Day','Month','Year'])
```

```python
result
```

```python
result['Datetime']=pd.to_datetime(result['Datetime'])
```

```python
result
```

```python
table = pd.pivot_table(result, values='Time', index=['person'],
                    columns=['Datetime','zone'], aggfunc='first')
```

```python
result.to_csv('write_april.csv')
```
